# Matplotlib example workflow

```python
#0.Import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1.Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2.Setup data
fig,ax=plt.subplots(figsize=(5,5)) #(width, height)

#3.Plot data
ax.plot(x,y)

#4.Customize plot
ax.set(title="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#5.Save & show (you save the whole figure)
fig.savefig("sample-plot.png")
```

### Scatter Plot

```python
fig ,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig , ax = plt.subplots()
ax.scatter (x,np.sin(x))
```

### Bar Plot

```python
# Make a plot from dictionary 
nut_butter_price = {"Almond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter":12}
fig,ax=plt.subplots()
ax.bar( nut_butter_price.keys(), nut_butter_price.values())
ax.set(title = "Butter Sales",
      xlabel = "Butters",
      ylabel = "Prices");
```

```python
fig , ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()));
```

### Histogram Plot

```python
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist(x);
```

# Subplot method-1 

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2 , 
                                          ncols = 2 , 
                                          figsize= (10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(),nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

# Subplots Method-2

```python
fig , ax = plt.subplots(nrows = 2 ,
                       ncols = 2,
                       figsize= (10,5))

ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(),nut_butter_price.values());4
ax[1,1].hist(np.random.randn(1000));
```

# Plotting from Pandas dataframe

```python
import pandas as pd
car_sales =pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index =pd.date_range ("1/1/2020",  periods = 1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("$","",regex = False)
car_sales["Price"] = car_sales["Price"].str.replace(",","",regex = False)
car_sales
```

```python
car_sales["Sales Date"]=pd.date_range("1/1/2020",periods = len(car_sales))
car_sales
```

```python
car_sales["Total Sales"]= car_sales["Price"].astype(float).cumsum()
car_sales
```

```python
car_sales.plot(x="Sales Date",y="Total Sales");
```

```python
# Reassign price column to int
car_sales["Price"] = car_sales["Price"].astype(float)

# Plot scatter plot with price column as numeric
car_sales.plot(x="Odometer (KM)",y ="Price",kind = "scatter");
```
